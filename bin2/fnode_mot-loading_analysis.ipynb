```python
%matplotlib notebook
from QDP import qdp
import os.path
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
dp = '/home/ebert/jupyter-notebooks/FNode-data'
exp_name = '2018_01_09_13_04_26_cs-mot-Ipgc-scan'
exp_date = '_'.join(exp_name.split('_')[:3])
experiment_file = os.path.join(exp_date, exp_name, 'results.hdf5')
q = qdp.QDP(base_data_path=dp)
q.load_data_file(experiment_file)
```

```python
q.experiments[0]['variable_list']
```

```python
means = []
means2 = []
means_norm = []
for i in q.experiments[0]['iterations'].keys():
    sig_data = q.experiments[0]['iterations'][i]['signal_data']
    means.append(np.mean(sig_data, axis=0)[0])
    means2.append(np.mean(sig_data, axis=0)[1])
    means_norm.append(means2[-1]/means[-1])
```

```python
f = plt.figure()
plt.plot(means)
```

```python
f = plt.figure()
plt.plot(means_norm)
```
